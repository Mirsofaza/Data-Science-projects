```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\CS\Downloads\Compressed\Introduction to Data Science Assignment2\Topic_Survey_Assignment.csv')
```

```python
df
```

```python
df=df.sort_values(by=['Very interested'],ascending=False)
```

```python
df['Very interested']=round(df['Very interested']/2233,2)
df['Somewhat interested']=round(df['Somewhat interested']/2233,2)
df['Not interested']=round(df['Not interested']/2233,2)
```

```python
plt.bar(df['Unnamed: 0'],df['Very interested'])
plt.title("Very interested")
plt.xlabel("Name")
plt.ylabel("Very interested")
plt.xticks(rotation=45)
```

```python
plt.bar(df['Unnamed: 0'],df['Somewhat interested'])
plt.title("Somewhat interested")
plt.xlabel("Name")
plt.ylabel("Somewhat interested")
plt.xticks(rotation=45)
```

```python
plt.bar(df['Unnamed: 0'],df['Not interested'])
plt.title("Not interested")
plt.xlabel("Name")
plt.ylabel("Not interested")
plt.xticks(rotation=45)
```
