```python
import pandas  as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
with open(r'C:\Users\CS\Downloads\Compressed\Introduction to Data Science Assignment2\wiki-Vote.txt\Wiki-Vote.txt') as f:
    lines = f.readlines()
```

```python
lines[1].split('\t')
```

```python
_from=[]
_to=[]

for i in lines[1:]:
    x=i.split('\t')
    _from.append(int(x[0]))
    _to.append(int(x[1]))
```

```python
G = nx.Graph()


for i in range(1000):
  a=_from[i]
  b=_to[i]
  G.add_edge(a,b)

nx.draw_networkx(G)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
plt.show()
```
